```python
def estimate_model(model, x, y):
    '''
    Performs 5-fold cross validation on model and returns dictionary with run time info & scoring metric info
       model - Model to be run and estimated
       x - matrix input independent variable observations
       y - matrix of output dependent variable observations
    '''
    cv = KFold(n_splits=5, random_state=42, shuffle=True)
    return cross_validate(model,
                          x,
                          y,
                          cv=cv,
                          scoring=('r2','neg_mean_absolute_error', 'neg_mean_squared_error'))
```

```python
def get_means(results_dictionary):
    '''
    Returns average of all keys inside dictionary. 
    results_dictionary must have keys with numeric values
    '''
    if type(results_dictionary) != dict: return None
    average_dict = {}
    for key, val in results_dictionary.items():
        key += ' average'
        average_dict[key] = val.mean()
    return average_dict
```

```python
def print_dictionary_elts(dictionary):
    '''
    Pretty print key/value pairs of a dictionary
    '''
    for key,val in dictionary.items():
        print(f'{key}: {val}')
```

```python
def get_predictions(model, independent):
    '''
    Returns model predictions based on independent variables passed as an argument.
    '''
    return model.predict(independent)
```

```python
def get_pred_errors(model, independent, actual_dependent):
    '''
    Returns MSE and MAE of predictions from the specified model
    '''
    predictions = get_predictions(model, independent)
    error_dict = {}
    error_dict['MSE'] = \
        metrics.mean_squared_error(actual_dependent, predictions, multioutput='raw_values')
    error_dict['MAE'] = \
        metrics.mean_absolute_error(actual_dependent, predictions, multioutput='raw_values')
    return error_dict
```

```python
def get_difference(actual, predicted):
    '''
    Returns the difference actual-predicted
    '''
    return actual - predicted
```
